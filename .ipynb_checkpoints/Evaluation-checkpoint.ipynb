# Important Note 

When handling categorical variables, it is **not** okay to just set them as 1,2,3...n values, as that would implement a constraint / bias that 1 is better than 2 etc. So, need to transform all event types into their individual category column binaries. This is done by scikit-learn too.

## Therefore, option 1 : discretize ALL

* Event type : 525 columns
* Gender : 2 columns
* Year : 2016,2015,2014,2013,2012 - 5 column
* Location : 232
* Event Name ? : 582
* Age : Discretize depending on category. below 15, 15-25,25-35 etc
* Discretize ID's : 8000+ columns

## Discretizing data

```python
def discretize(df,columns):
    COLS = []
    for i,column in enumerate(columns):
        print 'Checking ' + column
        df[column].fillna('UNK'+column,inplace=True)
        cf = df[column].value_counts().reset_index()
        chs = [row['index'] for i,row in cf.iterrows()]
        #print i
        #print chs
        COLS.extend(chs)
    #COLS = list(itertools.chain.from_iterable(COLS))
    #disc_df = pd.DataFrame(columns=COLS)
    full = []
    for i,row in df.iterrows():
        r = np.zeros(len(COLS))
        for column in columns:
            r[COLS.index(row[column])] = 1
        full.append(r)
    print 'Agg done'
    full = np.vstack(full)
    print 'vstack done'
    #disc_df = disc_df.append(pd.DataFrame(np.array([r]),columns=COLS),ignore_index=True)
    disc_df = pd.DataFrame(full,columns=COLS)
    return disc_df
```

```python
def transform(df,COLS):
    return discretize(df,COLS)
```

## Aggregating Data for Classification

#### Years in which user has participated in a Montreal Marathon (0 no, 1 yes)

```python
years = [2012,2013,2014,2015]

def transformYearData(num_users):
    # Calculating Year
    ndr = []
    for x in range(num_users):
        nr = np.zeros(len(years)) # 2012, 2013, 2014, 2015
        for i,rec in mon_mara[mon_mara['PARTICIPANT_ID'] == x].iterrows():
            nr[years.index(rec['YEAR'])] = 1
        ndr.append(nr)
    return np.vstack(ndr)
```

#### Total Montreal Marathons participated by a user
Also accounts for _all_ marathons in Montreal (including demi,half etc, which were labelled LABEL = 1)

```python
def transformTotalMaraMon(num_users):
    dt = data[(data.LABEL == 1) & data.LOCATION.str.contains('montreal',case=False)].groupby(['PARTICIPANT_ID']).agg('count').reset_index()
    numr = []
    for x in range(num_users):
        v = dt[dt.PARTICIPANT_ID == x]['EVENT_DATE']
        if len(v) > 0:
            numr.append(v.iloc[0])
        else:
            numr.append(0)
    return numr
```

```python
len(ndr_total_m)
ndr_total_m = np.array(ndr_total_m).reshape(-1,1)
```

#### Total Events participated by a user

```python
def transformTotal(num_users):
    dt = data.groupby(['PARTICIPANT_ID']).agg('count').reset_index()
    numr = []
    for x in range(num_users):
        v = dt[dt.PARTICIPANT_ID == x]['EVENT_DATE']
        if len(v) > 0:
            numr.append(v.iloc[0])
        else:
            numr.append(0)
    return numr
```

#### Total Events Participated by User in each year

```python
years = [2012,2013,2014,2015,2016]
def transformYearTotalData(num_users):
    # Calculating Year
    ndr = []
    dt = data.groupby(['PARTICIPANT_ID','YEAR']).agg('count').reset_index().sort_values('EVENT_DATE',ascending=False)
    for x in range(num_users):
        nr = np.zeros(len(years)) # 2012, 2013, 2014, 2015 and 2016
        for i,rec in dt[dt['PARTICIPANT_ID'] == x].iterrows():
            nr[years.index(rec['YEAR'])] = rec['EVENT_DATE']
        ndr.append(nr)
    return np.vstack(ndr)
```

```python
genderMap = {'M' : 0, 'F' : 1}
def transformGender(num_users):
    ndr = []
    for x in range(num_users):
        done = False
        for i,rec in data[data['PARTICIPANT_ID'] == x].iterrows():
            if rec['GENDER'] == 'M' or rec['GENDER'] == 'F':
                ndr.append(genderMap[rec['GENDER']])
                done = True
                break
        if not done:
            ndr.append(-1)
    return ndr
```
